```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
gas = pd.read_csv('gas_prices.csv')
gas
```

```python
plt.title('Valor do combustivel em US$')

list_country = ['Australia','Ytaly','Germany','Japan']

for country in gas:
    if country in list_country:
        plt.plot(gas['Year'], gas[country], label=[country], marker='.')
    

plt.xlabel('Years')
plt.ylabel('US$')
plt.xticks(gas['Year'][::2])
plt.legend()
plt.savefig('country.png')
plt.show()
```
